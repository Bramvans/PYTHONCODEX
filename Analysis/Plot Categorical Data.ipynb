## Scatterplot

```python
sns.catplot(x="species", y="sepal_length", kind = "swarm", data=iris)

# Opties

# - Swarm < - plot points also in the width of the graph for visual distribution
# - Jitter <- jitter the dots or keep them straight
# - Hue <- Choose an color for a subset in the category (eg m/f)
# - order <- pass a list of the order of the X axis
# - HORIZONTALPLOT <- just swap x and y axis
```

# Add Summary Metrics

```python
mean = iris['sepal_length'].mean() #<-- create the summary Metric
median = iris['sepal_length'].median() #<-- create the summary Metric
```

```python
import matplotlib.pyplot as plt 

# V line for vertical lines
# H line for horizontal lines

sns.catplot(x="species", y="sepal_length", kind = "swarm", data=iris)
plt.axhline(mean, color='r', linestyle='--')
plt.axhline(median, color='g', linestyle='-')
plt.legend({'Mean':mean,'Median':median})
```
